```python
import matplotlib.pyplot as plt
```

```python
x_value = [0,1,2,3,4,5,6]
y_value = [0,1,4,6,9,11,19]

plt.plot(x_value,y_value)
plt.show()
```

```python
x_value = [0,1,2,3,4,5]
y_value = [0,1,4,9,16,25]

plt.plot(x_value,y_value)
plt.text(1,1,"SMIT")
plt.title("DATA VISUALIZATION")
plt.xlabel("x label")
plt.ylabel("y label")
plt.savefig("smitfig.png")
plt.show()
```

```python
x_value = [0,1,2,3,4,5]
y_value = [0,1,4,9,16,25]
y2_value = [2,4,6,12,18,27]

plt.plot(x_value,y_value,label='a')
plt.plot(x_value,y2_value,label='b')
plt.text(1,1,"SMIT")
plt.title("DATA VISUALIZATION")
plt.xlabel("x label")
plt.ylabel("y label")
plt.xlim(0,10)
plt.ylim(0,30)
#plt.savefig("smitfig.png")
plt.legend()
plt.show()
```
